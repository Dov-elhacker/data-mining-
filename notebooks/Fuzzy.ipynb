```python
age = ctrl.Antecedent(np.arange(20, 81, 1), 'age')
income = ctrl.Antecedent(np.arange(4000, 100001, 1000), 'income') 
loan_amnt = ctrl.Antecedent(np.arange(500, 35001, 500), 'loan_amnt')
person_credit_history = ctrl.Antecedent(np.arange(2, 31, 1), 'person_credit_history') 
loan_status = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'loan_status')
```

```python
age['young'] = fuzz.trapmf(age.universe, [20, 20, 25, 35]) 
age['middle_aged'] = fuzz.trimf(age.universe, [30, 45, 60]) 
age['senior'] = fuzz.trapmf(age.universe, [55, 70, 80, 80]) 

income['low'] = fuzz.trapmf(income.universe, [4000, 4000, 20000, 40000]) 
income['medium'] = fuzz.trimf(income.universe, [30000, 60000, 90000]) 
income['high'] = fuzz.trapmf(income.universe, [80000, 150000, 1000000, 1000000]) 

loan_amnt['small'] = fuzz.trapmf(loan_amnt.universe, [500, 500, 5000, 15000]) 
loan_amnt['medium'] = fuzz.trimf(loan_amnt.universe, [10000, 20000, 30000]) 
loan_amnt['large'] = fuzz.trapmf(loan_amnt.universe, [25000, 30000, 35000, 35000]) 

person_credit_history['short'] = fuzz.trapmf(person_credit_history.universe, [2, 2, 4, 6])
person_credit_history['medium'] = fuzz.trimf(person_credit_history.universe, [5, 10, 15])
person_credit_history['long'] = fuzz.trapmf(person_credit_history.universe, [12, 18, 30, 30])

loan_status['safe'] = fuzz.trapmf(loan_status.universe, [0, 0, 0.2, 0.4])
loan_status['medium'] = fuzz.trimf(loan_status.universe, [0.3, 0.5, 0.7])
loan_status['risky'] = fuzz.trapmf(loan_status.universe, [0.6, 0.8, 1, 1])
```

```python
rule1 = ctrl.Rule(income['low'] & loan_amnt['large'], loan_status['risky'])

rule2 = ctrl.Rule(person_credit_history['short'] & loan_amnt['large'], loan_status['risky'])

rule3 = ctrl.Rule(person_credit_history['short'] & income['low'], loan_status['risky'])

rule4 = ctrl.Rule(age['young'] & person_credit_history['short'], loan_status['risky'])

rule5 = ctrl.Rule(income['medium'] & loan_amnt['medium'], loan_status['medium'])

rule6 = ctrl.Rule(person_credit_history['medium'] & loan_amnt['large'], loan_status['medium'])

rule7 = ctrl.Rule(age['young'] & income['medium'], loan_status['medium'])

rule8 = ctrl.Rule(income['high'] & person_credit_history['long'], loan_status['safe'])

rule9 = ctrl.Rule(person_credit_history['long'] & loan_amnt['small'], loan_status['safe'])

rule10 = ctrl.Rule(income['medium'] & person_credit_history['long'], loan_status['safe'])

rule11 = ctrl.Rule(age['middle_aged'] & person_credit_history['long'], loan_status['safe'])
```

```python
sprinkler_system = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10, rule11])
```

```python
sprinkler_simulator= ctrl.ControlSystemSimulation(sprinkler_system)
```

```python
sprinkler_simulator.input['loan_amnt'] = 90000
sprinkler_simulator.input['income'] = 15000
sprinkler_simulator.input['age'] = 55
sprinkler_simulator.input['person_credit_history'] = 3
```

```python
sprinkler_simulator.compute()
```

```python
loan_status_value =sprinkler_simulator.output['loan_status']
```
